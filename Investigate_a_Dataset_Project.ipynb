```python
#Import libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your data and print out a few lines. Perform operations to inspect data
#   types and look for instances of missing or possibly errant data.
#loading the csv file and storing it in the variable "data"
data = pd.read_csv('tmdb-movies.csv')

#printing first five rows with defined columns of tmdb-movies database
data.head()
```

```python
def shape(x,y):
  print('Number of Columns is : {0} \nNumber of Rows is : {1}'.format(x,y))
shape(data.shape[1],data.shape[0]) 
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(labels = ['imdb_id','homepage', 'budget_adj', 'revenue_adj'],axis =1,inplace =True)
```

```python
data.shape
```

```python
#drop duplicate values
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
# view missing value count for each feature
data.isnull().sum()
```

```python
# drop rows with any null values.
data.dropna(inplace=True)
```

```python
sns.pairplot(data)
```

```python
c=data.corr()
sns.heatmap(c,annot=True)
```

```python
#plotting a histogram of runtime of movies
plt.xlabel('Runtime of the Movies')
plt.ylabel('Nos.of Movies in the Dataset')
plt.title('Runtime of all the movies')
#giving a histogram plot
plt.hist(data['runtime'], rwidth = 0.9, bins =35)
#displays the plot
plt.show()
```

```python
data.plot(x='popularity',y='revenue',kind='scatter',color='#787FF6')
plt.title('Popularity Vs Revenue')
plt.xlabel('Popularity')
plt.ylabel('Revenue');
```
